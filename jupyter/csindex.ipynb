```python
from selenium.webdriver import Chrome, ChromeOptions
```

```python
chrome_options = ChromeOptions()
chrome_options.add_argument('headless')  # 启用静默模式，不打开浏览器窗口
driver = Chrome(options=chrome_options)
```

```python
url = 'https://www.csindex.com.cn/#/indices/family/detail?indexCode=000300'
driver.get(url)
```

```python
hushen300 = driver.find_element_by_xpath('//*[@id="app"]/div/div/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div[3]/a')
download_url = hushen300.get_property('href')
download_url
```

```python
import pandas as pd
df = pd.read_excel(download_url, usecols=[4,5,9], names=['code', 'name', 'weight'], dtype={'code': str})
df.dtypes
```
